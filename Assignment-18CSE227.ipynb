```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = np.arange(-20,21)
n
```

```python
ar = []
for i in n:
    if i>=0 and i<=10:
        ar .append((i*i)+i)
    elif i>=-10 and i<=0:
         ar .append(i-(i*i))
    else:
         ar .append(0)
br = np.array(ar)
br
```

```python
plt.stem(n,br)
```

```python
br = []
for i in n:
    if(i+20-5>=0 and i+20+5<41):
     br.append(ar[i+20-5]*ar[i+20+5])
#print(br)
cr = np.array(br)
print(cr)
```
